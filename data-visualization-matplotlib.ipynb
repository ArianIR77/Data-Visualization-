```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.7,0.7,0.2,0.2])
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('line with slope of 2')
ax2.plot(x,z)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('x^2')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

ax2.plot(x,z)
ax2.set_xlim(40,60)
ax2.set_ylim(1000,6000)
ax2.set_title('zoom')
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y,color='red', lw=3 , ls='--')
ax[0].grid(True)
ax[1].plot(-x,y,color='green', lw=3,ls='--')
ax[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

```python
n = np.array([0,1,2,3,4,5])
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(x*x, x*x + 0.25*np.random.randn(len(x*x)),s=1)
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
axes[3].set_title("fill_between")
```
